```python
test = pd.read_csv('us events ad_revenue filtered 03.02-07.02.csv')
```

```python
df=test['event_json'].str.split(',')[0]
df
```

```python
import json
```

```python
#распарсим столбец event_json
test_df = pd.io.json.json_normalize(test.event_json.apply(json.loads))
```

```python
#соединим два df
final_df = pd.concat([test, test_df], axis=1)
```

```python
final_df.head()
```

```python
final_df.drop(columns=['event_json'], axis=1)
```

```python
final_df.info()
```

```python
final_df.describe()
```

```python
agg_func_describe = {'ecpm_sma': ['describe']}
final_df.groupby(['ad_type']).agg(agg_func_describe).round(2)
```

```python
pd.pivot_table(final_df, index=['ad_type'])
```

**Наблюдения **

В представленных выше таблицах видно, что больше всего было показано баннерных объявлений (кол-во = 81598), вторая по популярности полноэкранная реклама (кол-во = 37613), на последнем месте - видео (кол-во = 6482). Если сравнивать eCPM, то показатель выше всего у видео-рекламы, что говорит о том именно с видео приходит наибольшая выручка. Этот показатель практически в 2 раза выше, чем у полноэкранной рекламы и практически в 10 раз выше, чем у баннерной.
